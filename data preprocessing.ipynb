```python
import pandas as pd
import numpy as np
```

```python
df_north = pd.read_excel('data/Mien Bac.xlsx')
df_south = pd.read_excel('data/Mien Nam.xlsx')
df_central = pd.read_excel('data/Mien Trung.xlsx')
```

```python
df_north_filled = df_north.fillna('NA')
df_south_filled = df_south.fillna('NA')
df_central_filled = df_central.fillna('NA')
```

```python
def standardize_names(df): #mục đích thống nhất tên ngành học giữa các tập dữ liệu, giúp việc tích hợp dữ liệu và phân tích sau này được nhất quán và chính xác hơn.
    mapping = {
        'Máy tính và công nghệ thông tin': 'Công nghệ thông tin',
        'Nhóm ngành máy tính và công nghệ thông tin': 'Công nghệ thông tin',
        'Kỹ thuật máy tính': 'Công nghệ thông tin'
    }
    for col in df.columns:
        for key, value in mapping.items():
            if key in col:
                new_name = col.replace(key, value)
                df.rename(columns={col: new_name}, inplace=True)
```

```python
standardize_names(df_north_filled)
standardize_names(df_south_filled)
standardize_names(df_central_filled)

# Show some columns from the datasets to verify changes
df_north_filled.head(), df_south_filled.head(), df_central_filled.head()
```

### RESPONSES DATA


```python
df_res = pd.read_excel('data/Responses.xlsx')
```

```python
# Chuẩn hóa tên miền
domain_standardization = {
    'Miền Bắc': 'Bắc',
    'Miền Trung': 'Trung',
    'Miền Nam': 'Nam'
}
df_res['Bạn muốn học đại học ở khu vực nào?'] = df_res['Bạn muốn học đại học ở khu vực nào?'].map(domain_standardization)
df_res['Bạn có dự định học ngành nào không?'].fillna('NA')
```

```python
# One hot encoding
# Mã hóa dữ liệu phân loại bằng One-Hot Encoding
# Giả sử các cột kỹ năng cần được mã hóa
skills_columns = ['Kỹ năng máy tính và tin học văn phòng', 'Kỹ năng ngôn ngữ', 'Kỹ năng vận hành và sử dụng máy móc', 'Kỹ năng phân tích dữ liệu', 'Kỹ năng bán hàng, tiếp thị', 'Kỹ năng viết', 'Kỹ năng tài chính', 'Kỹ năng quản lý dự án', 'Kỹ năng y tế']
data_responses_encoded = pd.get_dummies(df_res, columns=skills_columns)
print(data_responses_encoded.head())
```

```python
# Đảm bảo tất cả các DataFrame có cùng tên cột
df_north.columns = df_south.columns = df_central.columns

all_uni = pd.concat([df_north, df_south, df_central], ignore_index=True)
# In ra một số hàng đầu tiên để kiểm tra
print(all_uni.head())

# Kiểm tra thông tin về DataFrame mới để xác định số lượng hàng và cột
print(all_uni.info())

```
